```python
%matplotlib inline
from math import exp, log
import matplotlib.pyplot as plt

# initialization
w = .6 # weight
b = .9 # bias
lr = 0.15 # learning rate
n_epochs = 300 # number of epochs

# training data
x = 1.
y = 0.

# activation function
def sigmoid(z):
    return 1. / (1. + exp(-z))

# cost function
def cost(w, b):
    a = sigmoid(w * x + b)
    return (a - y) ** 2 / 2

# derivative of a function
def get_derivative(func):
    def func_wrapper(name):
        if name == 'w':
            return (func(w + 10e-5, b) - func(w, b)) / 10e-5
        elif name == 'b':
            return (func(w, b) - func(w, b)) / 10e-5
    return func_wrapper

# derivative of sigmoid function
@get_derivative
def dsigmoid(w, b):
    return cost(w, b)

costs = []

# start training
for i in range(n_epochs):
    h = 10 ** -5
    w -= lr * dsigmoid('w')
    b -= lr * dsigmoid('b')
    costs.append(cost(w, b))
    
plt.plot(range(n_epochs), costs)
plt.xlabel('epoch')
plt.ylabel('cost')
plt.show()
```

```python
# change cost function to cross entropy
def cost(w, b):
    a = sigmoid(w * x + b)
    return -(y * log(a) + (1 - y) * log((1 - a)))

# derivative of sigmoid function
@get_derivative
def dsigmoid(w, b):
    return cost(w, b)

costs = []

# start training
for i in range(n_epochs):
    h = 10 ** -5
    w -= lr * dsigmoid('w')
    b -= lr * dsigmoid('w')
    costs.append(cost(w, b))
    
plt.plot(range(n_epochs), costs)
plt.xlabel('epoch')
plt.ylabel('cost')
plt.show()
```
